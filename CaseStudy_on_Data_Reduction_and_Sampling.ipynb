```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/Spotify_Youtube_Sample.csv')
df.head()
```

```python
df.info()
```

```python
sns.pairplot(data=df.loc[:10],hue='Album_type')
```

###Data Reduction

```python
df1=df.drop(df[['Stream','Views','Likes','Comments']],axis=1)
df1.info()
```

```python
sns.pairplot(data=df1.loc[:10],hue='Album_type')
```

###Sampling

```python
df2=df.sample(n=5)
df2
```

```python
sns.pairplot(data=df2)
```

```python
df3=df.sample(frac=0.001)
df3
```

```python
df.loc[:10].sample(15,replace=True)
```
